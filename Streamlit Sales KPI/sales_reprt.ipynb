```python
sales_perday = df.groupby(['so_create_date', 'monthnum'], as_index=False)[['transaction_price']].sum()
```

```python
month = str(datetime.datetime.now().month).zfill(2)
```

```python
now = sales_perday[sales_perday['monthnum'] == month][-3:]['so_create_date'].unique()[0]
yesterday = sales_perday[sales_perday['monthnum'] == month][-4:]['so_create_date'].unique()[0]
```

```python
salesno = sales_perday[sales_perday['so_create_date'] == now]['transaction_price'].sum()
```

```python
salesyes = sales_perday[sales_perday['so_create_date'] == yesterday]['transaction_price'].sum()
```

```python
delta = round((salesno / salesyes ) - 1, 3)
```
